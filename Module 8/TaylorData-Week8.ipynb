## Week 8 - __Eigenvalues and Eigenvectors__


### Let A be the matrix given below.

```python
import numpy as np

# Define the matrix A
A = np.array([[4, 5, 0], [-1, 0, 0], [0, -6, 1]])

# Define the vectors to check
v1 = np.array([1, 2, 3])
v2 = np.array([0, 1, 0])

# Function to check if a vector is an eigenvector
def is_eigenvector(A, v):
    Av = A @ v
    lambda_v = Av / v
    return np.allclose(lambda_v, lambda_v[0])
```

### Is the vector [1 2 3] an eigenvector of A? Verify your answer with a calculation in Python

```python
# Check if v1 is an eigenvector
is_v1_eigenvector = is_eigenvector(A, v1)
is_v1_eigenvector



```

### Is the vector [0 1 0] an eigenvector of A? Verify your answer with a calculation in Python

```python
# Check if v2 is an eigenvector
is_v2_eigenvector = is_eigenvector(A, v2)
is_v2_eigenvector
```

### Make a scatterplot of the user scores versus critics scores.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Load the dataset
data = pd.read_csv('video_game_data.csv')
data.describe()
```

```python

# Scatterplot of user scores vs critic scores
plt.scatter(data['critic_score'], data['user_score'])
plt.xlabel('Critic Score')
plt.ylabel('User Score')
plt.title('User Scores vs Critic Scores')
plt.show()
```

 
### On your scatterplot from part (a), sketch the approximate directions of the first and second principal components. This can be done by hand or in Python.

```python
# Perform PCA
pca = PCA(n_components=2)
pca.fit(data[['critic_score', 'user_score']])

# Get the principal components
components = pca.components_

# Plot the principal components
plt.scatter(data['critic_score'], data['user_score'])
plt.quiver(pca.mean_[0], pca.mean_[1], components[0, 0], components[0, 1], color='r', scale=5)
plt.quiver(pca.mean_[0], pca.mean_[1], components[1, 0], components[1, 1], color='g', scale=5)
plt.xlabel('Critic Score')
plt.ylabel('User Score')
plt.title('User Scores vs Critic Scores with Principal Components')
plt.show()
```
